# From collection module using Counter

Counter is used to count the repeated words and gives output in the form of dictinary or dict i.e. in key value pairs 

For an example z = 1,2,3,4,1,4,4,3,8,6,6,6,6,6,6,
with the help of Counter() we can count that how may time each number has came by passing z varaible in Counter() which gives output in dictionary form 

In Counter it provide most.common parameter which can gives most occurence or repeted counts of number form the variable z 

In Counter it also provide update parameter which helps to update the counts if some same number is add in the z variables

Below are the example 


```python
# importing  Counter from collection library

from collections import Counter

```

```python
# counting the repeated number in the variable 

z = [1,2,3,4,1,4,4,3,8,6,6,6,6,6,6,]
Counter(z)
```

```python
#Counting the most common occurence

z = [1,2,3,4,1,4,4,3,8,6,6,6,6,6,6,]
Counter(z).most_common(2)
```

```python
z = ['hey','okay' ,'hey','sleep' ,'smile','sleep']
Counter(z).most_common(1)
```

```python
# using update parameter to update list 

z = [1,2,3,4,1,4,4,3,8,6,6,6,6,6,6,]
ok = Counter(z)
ok.update([1,9,9,9,9,9,6,10]) # updating the value of z variable 
ok
```

# Example of Counter

```python
songss= pd.read_fwf('lyrics .txt',header = None)
songss
```

```python
# counting the most repeted words in songss varaible using  collection counter ()

from collections import Counter
occurence_letter = songss[0] #
jo = songss[1]
ku = Counter()

for i in occurence_letter:
    ku.update(i.split(';'))
he=[]
okas=[]

for t in ku.most_common(10):
    he.append(t[0])
    okas.append(t[1])


he.reverse()
okas.reverse()

plt.barh(he,okas)

plt.title("Most Popular Langu")
# plt.ylabel("Programming Languages")
plt.xlabel("Number of People Who Use")

plt.tight_layout()

plt.show()   
```

```python
# reading the csv file

data = pd.read_csv('databar.csv')
data.head()
```

```python
ids = data['Responder_id'] # fetching  Responder_id columns and putting in ids varaible 

lang_responses = data['LanguagesWorkedWith'] # fetching language.... columns and putting in leang- variable 

language_counter = Counter() # putting the Counter() in the language  

for response in lang_responses:
    language_counter.update(response.split(';'))

languages = []
popularity = []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

languages.reverse()
popularity.reverse()

plt.barh(languages, popularity)

plt.title("Most Popular Languages")
# plt.ylabel("Programming Languages")
plt.xlabel("Number of People Who Use")

plt.tight_layout() # Adjust the padding between and around subplots.



plt.show()
```

# Stackplot chart

Stack Plots are used to plot linear data, in a vertical order, stacking each linear plot on another. Typically, they're used to generate cumulative plots.

Basically we can comapre any specific time of periods by using stack plot




```python
#using the stack plot 

plt.figure(figsize=(10,6))

working_day = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday']

John_working_time = [2,3,5,4,2,8]

jimmy_working_time =[3,5,4,5,6,6]

Boult_working_time = [8,6,3,4,2,7]

hey =['John_working_time','Jimmy_working_time','Boult_working_time']

plt.stackplot(working_day,John_working_time,jimmy_working_time,Boult_working_time, labels=hey,edgecolor='black',linewidth=1)

plt.xlabel('Working Days of the Employees')
plt.ylabel('Working Time of the Employees')


plt.legend(loc='upper center')
plt.show()
```

# Filling the area between lines using fill_between()

Fill the area between two horizontal curves.

The curves are defined by the points (x, y1) and (x, y2). This creates one or multiple polygons describing the filled area.

You may exclude some horizontal sections from filling using where.

By default, the edges connect the given points directly. Use step if the filling should be a step function, i.e. constant in between x.


Basiccally it fill the ara between the curves.

Syntax is: matplotlib.pyplot.fill_between(x, y1, y2=0, where=None, interpolate=False, step=None, *, data=None, **kwargs)


```python
# reading the csv file

hey = pd.read_csv('fill between data.csv')
hey
```

```python
# storing the indivisual column for every column to plot 

ages =hey['Age']
all_developers =hey['All_Devs']
python = hey['Python']
java = hey['JavaScript']
```

```python
# ploting the graph with filling the area between the curves or gaps

plt.plot(ages,all_developers,color='black',linestyle='--',label='All_devs')

plt.plot(ages,python,color='green',linestyle='-',label='Python')

plt.plot(ages,java,color='magenta',linestyle='-.',label='Java')

plt.fill_between(ages,all_developers,python,where=(all_developers>python),interpolate=True,alpha=0.25)

plt.fill_between(ages,all_developers,where=(all_developers<=python),interpolate=True,alpha=0.25)


plt.legend()

plt.title('Salaries of diffrent software  Developer')
plt.xlabel('Ages of Employees')
plt.ylabel('Salaries of Employees')
plt.show()
```

## plt.GridSpec: More Complicated Arrangements

GridSpec is a feature in Matplotlib that allows you to create complex subplot layouts, providing more flexibility and control over the arrangement of subplots within a Figure.

It is particularly useful when you need to create plots with varying column and row spans or when you want to create a layout with different-sized subplots.

To use GridSpec, you need to import it from the matplotlib.gridspec module. Here's a basic syntax for creating a GridSpec:

matplotlib.gridspec.GridSpec(nrows, ncols, figure=None, left=None, bottom=None, right=None, top=None, wspace=None, hspace=None, width_ratios=None, height_ratios=None)

Basically first we have to create a gridspec for the plots then by using subplots we can plots the graph 


 this is an image of figure ,axis and etc 

```python
from matplotlib.gridspec import GridSpec
```

```python
# creating a grid of 2X2 for the ploting the figure and storing in the grid variable

grid = plt.GridSpec(2,2)
grid
```

Bassically plt.GridSpec() is used  to create a grid by passing ros and column below are the example of grid 

```python
#Ploting the subplot of the created grid 

plt.subplot(grid[0,0]);
plt.title('grid [0,0]')

# subploting grid for 0 rows and columns
plt.subplot(grid[0,1]);
plt.title('grid [0,1]')

plt.subplot(grid[1,0]);
plt.title('grid [1,0]')

plt.subplot(grid[1,1]);
plt.title('grid [1,1]')

plt.tight_layout()
```

In the above we have used the gridspec() to plot the  subplot 

```python
# Plotting the plots with the use of gridspec()

num = np.linspace(1,20,10)
mark = np.sin(num)

# ploting the size of figure then plots the num and marks data in the figure axes

pic = plt.figure(figsize=(10,10))

#using the  gridspec() to allocate ours axes

gd =  plt.GridSpec(2,2,wspace=0.60,hspace=0.70)

#ploting the axis usinng gridspec  ()

ax1 = pic.add_subplot(gd[0,0])
ax1.plot(num,mark);

#plotting axes 2 

ax2= pic.add_subplot(gd[0,1])
ax2.bar(num,mark)

#ploting axes 3

ax3= pic.add_subplot(gd[1,0])
ax3.plot(num,mark);

#ploting axes 4

ax4= pic.add_subplot(gd[1,1])
ax4.pie(num,mark);
```

In the above we have subploted the plot using grid i.e two rows and two columns from the .Bassicaly it take  postion of row and column then plots the graph

## Text inside a box

Passing a dictionary to the bbox argument of the text function the text will be displayed inside a box, as in the example below.

```python
#plotting text into the box using bbox funtion 

plt.figure(figsize=(10,5))
vc=[45,60,90,87]
look =np.sin(vc)
plt.plot(look);
plt.text(0.0,0.90,s='sin',c='#E17B65')
plt.text(1,-0.20,s='down' ,ha='center', color = 'Green', bbox={'facecolor':'black','pad':-1});
plt.text(2.0,0.75 ,'up',ha='center', color = 'magenta')
plt.text(3.0,-0.90 ,'lower',ha='center',color = 'red',bbox ={'facecolor':'orange','alpha':0.5});
```

## Annotations and arrows with annotate

Annotations are similar to basic texts, but the annotate function provides further parameters to annotate specific parts of the plot with arrows. If you dont want to use arrows probably your best choice is to use text

```python
# Ploting the graph using annotation

fig,ax=plt.subplots(figsize=(10,5))
ax.plot(np.sin(np.linspace(0,10)));
plt.annotate('blue for a sin',xy=(0,0),xytext=(0,-0.1))
ax.plot(np.cos(np.linspace(0,10)));
plt.annotate('green for a cos',xy=(1,1),xytext=(0,1.01));
```

Annotation is similar to the plt.text() but in annotation  we can plot arrowa easily by passing xy coordinates

```python
# marking the specific points on the  above plots

fig,ax=plt.subplots(figsize=(10,5))
ax.plot(np.sin(np.linspace(0,10)));
plt.annotate('blue for a sin',xy=(0,0),xytext=(-0.1,-0.75),arrowprops={'facecolor':'red', 'width':8,'headwidth':10})
ax.plot(np.cos(np.linspace(0,10)));
plt.annotate('green for a cos',xy=(1,1),xytext=(21,-0.25),arrowprops=dict(arrowstyle='->',connectionstyle="angle3,angleA=0,angleB=-90"));
```
